```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("CarPrice_Assignment.csv")
```

```python
df.info()
```

```python
ax = sns.relplot(x="carlength" , y= "price", data=df, palette="Set1")
```

```python
sns.relplot(x='carlength' , y='price', hue = "carheight", markers=["D", "o"],style="fueltype",palette="YlGnBu", data= df)
```

```python
sns.relplot(x='carlength' , y='price' , data= df, kind = "line")
```

```python
sns.set(style="darkgrid")
sns.countplot("fueltype",data=df)
sns.despine(trim= True)
```

```python
sns.catplot(x = "fueltype" , y="price" , data= df)
```

```python
sns.catplot(x = "fueltype" , y="price" , kind = "violin" , data= df)
```

```python
sns.catplot(x = "fueltype" , y="price" , kind = "box" , data= df)
```

```python
sns.set(style="darkgrid")
sns.catplot(x = "fueltype" , y="price"  , kind = "boxen" , data= df)
```

```python
df1 = df.iloc[:200, 20:]
df1
```

```python

g = sns.PairGrid(df1)
g.map(plt.scatter)
```

```python
sns.set(style="white" , color_codes = True)
sns.catplot(x = "fueltype" , y="price" , data= df)
sns.despine(trim= True)
```

```python
c = sns.color_palette()
sns.palplot(c)
```

```python
c = sns.hls_palette()
sns.palplot(c)
```

```python
c = sns.husl_palette()
sns.palplot(c)
```
