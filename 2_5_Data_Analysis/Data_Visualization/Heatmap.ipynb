```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/home/arraygen/Akshata/Machine_Learning/5_to_10_Machine_Learning/Linear_Regression_and_random_forest/insurance.csv")
```

```python
dataset.columns
```

```python
def smoker(yes):
    if yes =="yes":
        return 1
    else:
        return 0
dataset["smoker"]=dataset["smoker"].apply(smoker)

def sex(s):
    if s =="male":
        return 1
    else:
        return 0
dataset["sex"]=dataset["sex"].apply(sex)
```

```python
dataset.corr()["charges"]
```

```python
sns.heatmap(dataset.corr().abs())
```

```python
plt.figure(figsize=(10,7))
ax = sns.heatmap(dataset.corr().abs(), annot = True, cmap="YlGnBu")
```

```python
figure = ax.get_figure()   
figure.savefig("output.png")
```

```python

# Show the results of a linear regression within each dataset
sns.lmplot(x="age", y="charges", hue="sex", col ="smoker" ,data=dataset)
```
