## Line Plot

```python

# importing matplotlib module  
from matplotlib import pyplot as plt 
  
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot 
plt.plot(x,y) 
plt.title("Line Plot")
plt.xlabel("X-asix")
plt.ylabel("Y-axis")
  
# function to show the plot 
plt.show() 
```

```python
x1 = [1,2,3]
y1 =[5,3,8]
  
# Function to plot 
plt.plot(x,y,color ="r", label ="red color") 
plt.plot(x1,y1 , label ="blue color")
plt.title("Line Plot")
plt.xlabel("X-asix")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

## Bar plot

```python
plt.bar(x,y)
plt.bar(x1,y1)
plt.show()
```

## Historgram

```python
a = [22,87,55,43,56,73,55,54,11,20,51,5,79,31,27,99,77,67,20,78,45,78,89,2,33,43,22,21,23,78,87,55,44,6,44,55]
plt.hist(a, bins = [0,20,40,60,80,100],rwidth=0.985)
plt.title("histogram of result")
plt.xlabel('marks')
plt.ylabel('no. of students')
plt.show()
```

## Scatter Plot

```python
plt.scatter(x,y)
plt.scatter(x1,y1)
```

## Pie Chart

```python
slice_ =[7,2,2,13]
actvities = ['sleeping','eating','playing','Working']
plt.pie(slice_, labels = actvities, startangle = 90, shadow = True, explode =(0,0.1,0,0), autopct = "%1.1f%%")
plt.savefig("pie.png" , bbox_inches = "tight", pad_inches = 2 )
```
