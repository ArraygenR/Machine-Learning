# Data Types and Variables

## Variable Concept 

### Rules for variable name

1. You cannot start name with number but you can have number in between or at the end
2. You can't have special character(!@#$%^&*{}:";,.<>) in name except _
3. We can't have space in between 
4. Case sensitive
Ex : ABC = 2354
print(abc) # not allowed 
5. Can't have keyword as name.
Ex: if , else, for, while, def ,class

### 2. Arithmatic Operations

```python
num1 = 20
num2 = 3
```

```python
print("Number 1:", num1)
print("Number 2:", num2)
```

```python
add = num1 + num2
print(num1, "+", num2,"=", add)
```

```python
sub = num1 - num2
print(num1, "-", num2,"=", sub)
```

```python
nam1 = num1 / num2
print(num1, "/", num2, "=", nam1)
```
