# Control Flow Statement

It will help us to control flow of execeution

#### Relational operators used for conditions in control flow statement

<, >, var1 <= var2 ( var1 either less or equal than var2 then only True) , >= , != (2 things are not equal then only return True),  == (If both things are equal then only True)

## 1. Conditional Statement

```python
num1 = 10
if num1 > 0:
    print(num1, "is positive number")
```

```python
num1 = -10
if num1 > 0:
    print(num1, "is positive number")
```

```python
num1 = -10
if num1 > 0:
    print(num1, "is positive number")
else:
    print(num1, "is negative number")
```

```python
# for 0 this code will give wrong output
num1 = 0
if num1 > 0:
    print(num1, "is positive number")
else:
    print(num1, "is negative number")
```

```python
num1 = 0
if num1 > 0:
    print(num1, "is positive number")
elif num1 == 0:
    print(num1, "is not positive and not negative")
else:
    print(num1, "is negative number")
```

#### Q. Take input from user and check mwheather its positive or negative

```python
num1 = float(input("Enter any number:"))
if num1 > 0:
    print(num1, "is positive number")
elif num1 == 0:
    print(num1, "is not positive and not negative")
else:
    print(num1, "is negative number")
```
