# Image Processing

### 1. Using PIL

###### 1.1 Install pillow

###### 1.2 Import and read image

```python
#Import required library
import PIL
print('Pillow Version:', PIL.__version__)
```

```python
from PIL import Image

#Open Image
im = Image.open("data/BloodImage_00000.jpg")
im.show()

# summarize some details about the image
print(im.format)
print(im.mode)
print(im.size)
```

###### 1.3 Covert image to numpy array and vice versa 

```python
# load image and convert to and from NumPy array
from PIL import Image
from numpy import asarray
# load the image
image = Image.open('data/BloodImage_00000.jpg')
# convert image to numpy array
data = asarray(image)
# summarize shape
print(data.shape)
# create Pillow image
image2 = Image.fromarray(data)
# summarize image details
print(image2.format)
print(image2.mode)
print(image2.size)
#print(data) # 480 matrices , 640 rows in each matrix and 3 columns in each row
```

```python
#Image rotate & show
image.rotate(45).show()
```

```python
im_gray = Image.open('data/BloodImage_00000.jpg').convert('L')
im_gray.save("ImageCopy.jpg")
im_gray.show()
```

```python
img_resized = image.resize((200,200))
# report the size of the thumbnail
print(img_resized.size)
```

```python
img_resized.show()
```

### 2. Read by matplotlib

```python
# load and display an image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
data = image.imread('data/BloodImage_00000.jpg')
# summarize shape of the pixel array
print(data.dtype)
print(data.shape)
# display the array of pixels as an image
pyplot.imshow(data)
pyplot.show()
```

### 3. Using opencv

```python
#Import required library
import cv2
import numpy as np
from matplotlib import pyplot as plt

im = cv2.imread('data/BloodImage_00000.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('image',im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
